# Twitter Sentiment Analysis

```python
#import librarty
import stweet as st
#mencari kata yang ingin di cari
search_tweets_task = st.SearchTweetsTask(all_words='Cikarang')
#hasil Output
output_jl_tweets = st.JsonLineFileRawOutput('output_raw_search_tweets.jl')
output_jl_users = st.JsonLineFileRawOutput('output_raw_search_users.jl')
#
output_print = st.PrintRawOutput()

st.TweetSearchRunner(search_tweets_task=search_tweets_task,tweet_raw_data_outputs=[output_print, output_jl_tweets], user_raw_data_outputs=[output_print, output_jl_users]).run()
```

### Note
1. Library stweet adalah sebuah perpustakaan (library) open-source untuk mengakses dan melakukan pencarian data Twitter   menggunakan API Twitter
Dengan library stweet, pengguna dapat melakukan analisis data Twitter, seperti sentiment analysis, network analysis, dan topik trending. Library ini sangat berguna bagi para peneliti, jurnalis, dan pengguna lain yang tertarik dengan analisis data Twitter.

2. Bila ingin menganti topik hendaknya menganti nama file output atau menghapusnya kalau tidak akan menimpa

## Reading json Output to Dataframe

```python
import pandas as pd

tweet = pd.read_json('output_raw_search_tweets.jl', lines=True)

tweet.head()
```

```python
tweet['raw_value'][1]
```

```python
tweet_rapi = pd.json_normalize(tweet['raw_value'])
tweet_rapi['full_text'].head(10)
```

### Note
1. pd.read_json digunakan untuk membaca file JSON yang disimpan dalam direktori yang sama dengan file Python saat ini dan mengubahnya menjadi sebuah DataFrame
2. untuk mengimport data dari file JSON dengan struktur yang nested atau kompleks ke dalam program Python sebagai objek DataFrame yang memiliki struktur datanya flat.

## Scrapping on the specific period

```python
search_tweets_task = st.SearchTweetsTask(all_words='Cikarang until:2023-03-15 since:2023-03-01')

output_jl_tweets = st.JsonLineFileRawOutput('output_raw_search_tweets_1month.jl')
output_jl_users = st.JsonLineFileRawOutput('output_raw_search_users1month.jl')

output_print = st.PrintRawOutput()

st.TweetSearchRunner(search_tweets_task=search_tweets_task,tweet_raw_data_outputs=[output_print, output_jl_tweets], user_raw_data_outputs=[output_print, output_jl_users]).run()
```

```python
tweet_new = pd.read_json('output_raw_search_tweets_1month.jl', lines=True)

tweet_clean = pd.json_normalize(tweet_new.raw_value)
tweet_clean['created_at'] = pd.to_datetime(tweet_clean['created_at']) 

tweet_clean = tweet_clean[['full_text', 'created_at']]

tweet_clean.describe()
```

## Combining from another period

```python
search_tweets_task = st.SearchTweetsTask(all_words='Cikarang until:2023-02-27 since:2023-02-01')
output_jl_tweets = st.JsonLineFileRawOutput('output_raw_search_tweets_feb.jl')
output_jl_users = st.JsonLineFileRawOutput('output_raw_search_users_feb.jl')

output_print = st.PrintRawOutput()

st.TweetSearchRunner(search_tweets_task=search_tweets_task,
                     tweet_raw_data_outputs=[output_print, 
                                             output_jl_tweets], 
                     user_raw_data_outputs=[output_print, output_jl_users]).run()
```

```python
tweet_feb = pd.read_json('output_raw_search_tweets_feb.jl', lines=True)

tweet_clean_feb = pd.json_normalize(tweet_feb.raw_value)
tweet_clean_feb['created_at'] = pd.to_datetime(tweet_clean_feb['created_at']) 

tweet_clean_feb = tweet_clean_feb[['full_text', 'created_at']]

tweet_clean_feb.describe()
```

```python
tweet_join = pd.concat([tweet_clean_feb,tweet_clean])

tweet_join.describe()
```

## Creating Visualisation

```python
pd.crosstab(index = tweet_join['Day'] , 
            columns = 'count').sort_values(by = 'count', ascending =False)
```

```python
pd.crosstab(index = tweet_join['Day'] , 
            columns = 'count').sort_values(by = 'count', ascending =False).plot(kind = 'bar')
```

```python
tweet_join['date'] = pd.to_datetime(tweet_join['created_at'].dt.date)

pd.crosstab(index = tweet_join['date'], columns='count').plot()
```

## Creating wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#filter the df to one candidate, and create a list of responses from them
text = tweet_join['full_text'].tolist() 

# join the list and lowercase all the words
text = ' '.join(text).lower()

#create the wordcloud object
wordcloud = WordCloud(collocations=True).generate(text)

#plot the wordcloud object
plt.imshow(wordcloud, interpolation='bilInear')
plt.axis('off')
plt.show()
```
